### Monitor de Temperatura do Sistema

Este script simula a verificação da temperatura do sistema. Para leituras reais da temperatura do sistema, seriam necessárias bibliotecas específicas do SO (por exemplo, `psutil` no Linux/Windows, `osx-cpu-temp` no macOS, ou acesso direto ao hardware).

```python
def verificar_temperatura(temperatura_atual):
    """
    Verifica a temperatura do sistema e exibe um alerta se estiver muito alta.

    Args:
        temperatura_atual (float): A temperatura atual do sistema.
    """
    print(f"Temperatura atual do sistema: {temperatura_atual}°C")
    if temperatura_atual > 30:
        print("ALERTA: Superaquecimento do sistema! A temperatura está acima de 30°C.")
    else:
        print("A temperatura do sistema está normal.")

# --- Demonstração ---
# Você pode alterar este valor para testar o alerta
temperatura_simulada = 35.5
# Ou, para uma temperatura normal:
# temperatura_simulada = 25.0

verificar_temperatura(temperatura_simulada)
```
