### Monitor de Temperatura do Sistema

Este script simula a verificação da temperatura do sistema. Para leituras reais da temperatura do sistema, seriam necessárias bibliotecas específicas do SO (por exemplo, `psutil` no Linux/Windows, `osx-cpu-temp` no macOS, ou acesso direto ao hardware).

```python
def verificar_temperatura(temperatura_atual):
    """
    Verifica a temperatura do sistema e exibe um alerta se estiver muito alta.

    Args:
        temperatura_atual (float): A temperatura atual do sistema em graus Celsius.
    """
    # Exibe a temperatura atual do sistema, usando concatenação de strings.
    print("Temperatura atual do sistema: " + str(temperatura_atual) + "°C")

    # Verifica se a temperatura atual excede o limite de 30 graus Celsius.
    if temperatura_atual > 30:
        # Se a temperatura for superior a 30°C, exibe um alerta de superaquecimento.
        print("ALERTA: Superaquecimento do sistema! A temperatura está acima de 30°C.")
    else:
        # Caso contrário, informa que a temperatura está normal.
        print("A temperatura do sistema está normal.")

# --- Demonstração do Uso ---
# Esta seção demonstra como usar a função 'verificar_temperatura'.
# Agora, você pode inserir a temperatura manualmente.

# Solicita ao usuário para inserir a temperatura.
temperatura_digitada_str = input("Por favor, digite a temperatura atual (em °C): ")

# Tenta converter a entrada para float, lidando com possíveis erros de valor.
try:
    temperatura_simulada = float(temperatura_digitada_str)
    # Chama a função para verificar a temperatura inserida.
    verificar_temperatura(temperatura_simulada)
except ValueError:
    print("Entrada inválida. Por favor, digite um número para a temperatura.")
```
