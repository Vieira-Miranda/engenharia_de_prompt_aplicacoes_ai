### 1. Configuração Inicial e Criação de Diretórios

Primeiro, vamos importar as bibliotecas necessárias (`os` para interagir com o sistema operacional e `shutil` para operações de arquivo de alto nível) e definir os diretórios de origem e destino. O diretório de destino será criado se ainda não existir.

```python
import os
import shutil

# Define o diretório de origem onde os arquivos bagunçados estarão
source_dir = 'arquivos_desorganizados'
# Define o diretório de destino onde as pastas organizadas serão criadas
destination_dir = 'arquivos_organizados'

# Cria o diretório de destino se ele não existir
if not os.path.exists(destination_dir):
    os.makedirs(destination_dir)
    print(f"Diretório de destino '{destination_dir}' criado.")
else:
    print(f"Diretório de destino '{destination_dir}' já existe.")

# Cria o diretório de origem se ele não existir (para os arquivos de exemplo)
if not os.path.exists(source_dir):
    os.makedirs(source_dir)
    print(f"Diretório de origem '{source_dir}' criado.")
else:
    print(f"Diretório de origem '{source_dir}' já existe.")
```

### 2. Geração de Arquivos Fictícios para Demonstração

Para que o programa possa ser testado, vamos criar 10 arquivos de cada tipo (PDF, JPG, XLS, DOC, PY, IPYNB) no diretório de origem. Isso simula uma situação de arquivos "bagunçados".

```python
file_extensions = ['.pdf', '.jpg', '.xls', '.doc', '.py', '.ipynb']
num_files_per_type = 10

print(f"\nGerando {num_files_per_type} arquivos fictícios para cada extensão em '{source_dir}'...")
for ext in file_extensions:
    for i in range(num_files_per_type):
        file_name = f"arquivo_exemplo_{i}{ext}"
        file_path = os.path.join(source_dir, file_name)
        with open(file_path, 'w') as f:
            f.write(f"Conteúdo de {file_name}")
        print(f"  Criado: {file_name}")
print("Geração de arquivos fictícios concluída.")
```

### 3. Definição das Categorias e Criação das Pastas de Destino

Vamos definir um mapeamento de extensões para nomes de pastas e criar essas pastas dentro do diretório de destino.

```python
# Mapeamento de extensões para nomes de pastas
file_categories = {
    '.pdf': 'Documentos_PDF',
    '.jpg': 'Imagens_JPG',
    '.jpeg': 'Imagens_JPG',
    '.png': 'Imagens_PNG',
    '.gif': 'Imagens_GIF',
    '.xls': 'Planilhas_XLS',
    '.xlsx': 'Planilhas_XLSX',
    '.doc': 'Documentos_DOC',
    '.docx': 'Documentos_DOCX',
    '.txt': 'Textos',
    '.py': 'Scripts_Python',
    '.ipynb': 'Notebooks_Jupyter',
    '.zip': 'Arquivos_Compactados',
    '.rar': 'Arquivos_Compactados'
}

print("\nCriando pastas de categorias no diretório de destino...")
for folder_name in set(file_categories.values()): # Usa set para evitar duplicatas de pastas
    folder_path = os.path.join(destination_dir, folder_name)
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"  Pasta '{folder_name}' criada.")
    else:
        print(f"  Pasta '{folder_name}' já existe.")
print("Criação de pastas de categorias concluída.")
```

### 4. Organização e Movimentação dos Arquivos

Agora, o programa irá iterar sobre cada arquivo no diretório de origem, identificar sua extensão e movê-lo para a pasta correspondente no diretório de destino.

```python
print("\nIniciando a organização e movimentação dos arquivos...")
# Lista todos os arquivos no diretório de origem
for filename in os.listdir(source_dir):
    source_path = os.path.join(source_dir, filename)

    # Verifica se é um arquivo (e não um subdiretório)
    if os.path.isfile(source_path):
        # Extrai o nome base e a extensão do arquivo
        name, ext = os.path.splitext(filename)
        ext = ext.lower() # Converte a extensão para minúsculas para padronização

        # Verifica se a extensão está mapeada para uma categoria
        if ext in file_categories:
            folder_name = file_categories[ext]
            destination_path = os.path.join(destination_dir, folder_name, filename)
            # Move o arquivo para a pasta de destino
            shutil.move(source_path, destination_path)
            print(f"  Movido: '{filename}' para '{folder_name}/'")
        else:
            # Se a extensão não for reconhecida, move para uma pasta 'Outros'
            other_folder = os.path.join(destination_dir, 'Outros')
            if not os.path.exists(other_folder):
                os.makedirs(other_folder)
            shutil.move(source_path, os.path.join(other_folder, filename))
            print(f"  Movido: '{filename}' para 'Outros/' (extensão não reconhecida).")
    else:
        print(f"  Ignorado (não é um arquivo): '{filename}'")

print("Organização de arquivos concluída!")
```

### 5. Verificação dos Resultados

Para confirmar que a organização funcionou, vamos listar o conteúdo do diretório de destino.

```python
print(f"\nConteúdo do diretório de destino '{destination_dir}':")
for root, dirs, files in os.walk(destination_dir):
    level = root.replace(destination_dir, '').count(os.sep)
    indent = ' ' * 4 * (level)
    print(f"{indent}{os.path.basename(root)}/")
    subindent = ' ' * 4 * (level + 1)
    for f in files:
        print(f"{subindent}{f}")

# Opcional: Você pode querer remover os diretórios criados após a demonstração
# shutil.rmtree(source_dir)
# shutil.rmtree(destination_dir)
# print(f"\nDiretórios '{source_dir}' e '{destination_dir}' foram removidos.")
```
