```python
def calculadora():
    """Função principal que implementa uma calculadora simples com as quatro operações básicas."""

    print("Selecione a operação:")
    print("1 - Soma")
    print("2 - Subtração")
    print("3 - Multiplicação")
    print("4 - Divisão")

    # Solicita a opção do usuário para a operação desejada
    escolha = input("Digite a opção (1/2/3/4): ")

    # Verifica se a opção selecionada é válida
    if escolha in ('1', '2', '3', '4'):
        try:
            # Solicita a entrada dos dois números para a operação
            num1 = float(input("Digite o primeiro número: "))
            num2 = float(input("Digite o segundo número: "))
        except ValueError:
            # Trata o erro caso a entrada não seja um número
            print("Entrada inválida. Por favor, digite um número.")
            return

        # Realiza a operação de acordo com a escolha do usuário
        if escolha == '1':
            print(f"Resultado: {num1} + {num2} = {num1 + num2}")
        elif escolha == '2':
            print(f"Resultado: {num1} - {num2} = {num1 - num2}")
        elif escolha == '3':
            print(f"Resultado: {num1} * {num2} = {num1 * num2}")
        elif escolha == '4':
            # Verifica se o segundo número não é zero antes da divisão
            if num2 != 0:
                print(f"Resultado: {num1} / {num2} = {num1 / num2}")
            else:
                print("Erro: Divisão por zero!")
    else:
        # Mensagem para opção inválida
        print("Opção inválida")

# Executa a função da calculadora
calculadora()
```
